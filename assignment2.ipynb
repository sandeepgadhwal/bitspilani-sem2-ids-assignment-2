### Prepare Skills List
The skills are in text format we need to extract skills in form of a list for each row for easy comaprision.

```python
stop_words = set(stopwords.words("english"))
def harmonize_column_text(value):
    a = re.sub('\s+', ' ', re.sub('[^a-zA-z\s]', '', value.lower())) 
    a = [x for x in nltk.word_tokenize(a) 
                   if not x in stop_words]
    lemma = WordNetLemmatizer()
    a = [lemma.lemmatize(x) for x in a]
    a =" ".join(a)
    return a
```

```python
df['preferred_skills_harmon'] = df.loc[~df['Preferred Skills'].isna(), 'Preferred Skills'].apply(harmonize_column_text)
```

### Extract Experience years from text

```python
df['experience_years'] = df['Minimum Qual Requirements'].str.extract('(\d+)\syear.+experience').astype(float)
```

```python
_experience_years = df['Minimum Qual Requirements'].str.extract('(\w+)\syear.+experience')[0]
```

```python
_experience_months = df['Minimum Qual Requirements'].str.extract('(\w+)\smonth.+experience')[0]
```

```python
_experience_years[df['experience_years'].isna()].unique()
```

```python
exceptions = {
    'half': 0.5,
    'a': 1,
    '8th': 8,
    'more': 0
}
```

```python
def cast(value):
    try:
        return w2n.word_to_num(value)
    except:
        return exceptions[value]
```

```python
df.loc[df['experience_years'].isna(), 'experience_years'] = _experience_years.loc[df['experience_years'].isna() & ~_experience_years.isna()].apply(cast)
```

```python
_experience_months[df['experience_years'].isna()].unique()
```

```python
df.loc[df['experience_years'].isna(), 'experience_years'] = _experience_months.loc[df['experience_years'].isna() & ~_experience_months.isna()].apply(cast)/12
```

## 5. Missing value or data

### Find columns with no data

```python
def plot_na(df):
    store = []
    for col in df.columns:
        nodata_r = df[col].isna().sum()*100/len(df)
        if nodata_r > 0:
            store.append({
                "Column": col,
                "% Nodata":  nodata_r
            })
            print(f"-- Column: {col} | % Nodata: {(nodata_r):0.2f} %")           
    if len(store) > 0:
        pd.DataFrame(store)\
        .sort_values('% Nodata', False)\
        .plot(kind='bar',x='Column', y='% Nodata')
    else:
        print("-- supplied dataframe has no NoData values")
    
```

```python
plot_na(df)
```

### Grouping 'Job Category' after cleaning a bit

```python
def get_job_cats(value):
    store = []
    for x in value.split(','):
        store.extend(x.split('&'))
    store = [x.strip() for x in store]
    store = [x for x in store if x!='']
    return store
```

```python
fdf['Job Cats'] = fdf['Job Category'].apply(get_job_cats)
```

```python
store = []
for _, row in fdf.iterrows():
    for x in row['Job Cats']:
        store.append({
            'Job Cat': x,
            'Salary Range From': row['Salary Range From'], 
            'Salary Range To': row['Salary Range To'] 
        })
#
jdf = pd.DataFrame(store)
jdf.head()
```

#### Find Median upper edge of Salary Range 

Here we are taking the median of 'Salary Range To' for all the jobs matching with the respective skills.

```python
skill_df['maxSalaryMin'] = 0.0
skill_df['maxSalaryMax'] = 0.0
skill_df['maxSalaryMedian'] = 0.0
blocked = ['skill', 'ability work', 'experience working', 'skill experience', 'year experience', 'new york', 'posse following', 'candidate posse', 'preferred candidate', 'york city', 'minimum year']
for i, row in skill_df.iterrows():
    a = False
    for b in blocked:
        if row.Skill[:len(b)] == b:
            a = True
            continue
    if a: continue
    mask = fdf.preferred_skills_harmon.str.contains(row.Skill)
#     skill_df.loc[i, 'minSalary'] = _sdf.loc[mask, 'Salary Range From'].min()
    skill_df.loc[i, 'maxSalaryMin'] = fdf.loc[mask, 'Salary Range To'].min()
    skill_df.loc[i, 'maxSalaryMax'] = fdf.loc[mask, 'Salary Range To'].max()
    skill_df.loc[i, 'maxSalaryMedian'] = fdf.loc[mask, 'Salary Range To'].median()
```

```python
_sdf = skill_df.groupby('Skill') \
.first()[['maxSalaryMedian']] \
.sort_values('maxSalaryMedian').tail(10)
#
_sdf.plot(kind='barh', figsize=(8, 8))
```

## Q2 What are the job categories, which involve above mentioned niche skills?

```python
for i, row in skill_df.iterrows():
    a = False
    for b in blocked:
        if row.Skill[:len(b)] == b:
            a = True
            continue
    if a: continue
    mask = fdf.preferred_skills_harmon.str.contains(row.Skill)
#     skill_df.loc[i, 'minSalary'] = _sdf.loc[mask, 'Salary Range From'].min()
    skill_df.loc[i, 'maxSalaryMin'] = fdf.loc[mask, 'Salary Range To'].min()
    skill_df.loc[i, 'maxSalaryMax'] = fdf.loc[mask, 'Salary Range To'].max()
    skill_df.loc[i, 'maxSalaryMedian'] = fdf.loc[mask, 'Salary Range To'].median()
```

```python
store = {}
for skill in _sdf.index:
    for jobcats in fdf[fdf.preferred_skills_harmon.str.contains(skill)]['Job Cats']:
        for j in jobcats:
            if j not in store:
                store[j] = {
                    'Job Category': j,
                    'Count': 1,
                    'Skill': skill,
                    'Skills': [skill]
                }
            if skill != store[j]['Skill']:
                store[j]['Count']+=1
                store[j]['Skill'] = skill
                store[j]['Skills'].append(skill)
#
job_skill_df = pd.DataFrame(store.values())
job_skill_df.head()
```

```python
_sjdf = job_skill_df[['Skill']].reset_index().groupby('Skill').count()
_sjdf.columns = ['N Job Categorys']
_sjdf = _sjdf.sort_values('N Job Categorys')
_sjdf.plot(kind='barh', figsize=(12, 8))
```
